## Matplotlib 图像解剖图（Anatomy of a Figure）

1. Figure（图形区域）:
- plt.figure：整个图形的外框区域，包括所有的内容。是画布，所有的轴（axes）都会绘制在这里。
2. Axes（轴区域/子图）:
- fig.subplots：子图的实际区域，包含绘图的主要内容，比如网格、数据曲线等。一个 Figure 可以包含多个 Axes，每个 Axes 是一个单独的子图。
3. Title（标题）:
- ax.set_title：设置图表的标题，显示在子图的顶部。
4. xLabel / yLabel（轴标签）:
- ax.set_xlabel：设置 X 轴的名称。ax.set_ylabel：设置 Y 轴的名称。
5. xAxis / yAxis（轴对象）:
- ax.xaxis：X 轴的对象，可以设置刻度定位器（locator）、格式化器（formatter）等。ax.yaxis：Y 轴的对象，功能类似。
6. Major tick / Minor tick（主刻度/次刻度）: 
- 主刻度：较大的刻度，通常用于显示重要的值。
- ax.xaxis.set_major_locator：设置主刻度的位置。ax.xaxis.set_major_formatter：设置主刻度的格式。
- 次刻度：更小的刻度，用于更精确的刻度显示。
- ax.xaxis.set_minor_locator：设置次刻度的位置。ax.xaxis.set_minor_formatter：设置次刻度的格式。
7. Legend（图例）:
- ax.legend：图例用于说明数据的分类或含义，通常显示线条或标记的名字。
8. Line（线条）:
- ax.plot：绘制的折线或曲线。
9. Markers（标记）:
- ax.scatter：点标记（scatter plot）。
10. Grid（网格线）:
- ax.grid：图表上的网格线，用于辅助观察数据分布。
11. Spine（轴框）:
- ax.spines：子图的边框，可以单独设置四边的样式。

**matlab-syntax in matplotlib**

```python
import matplotlib.pyplot as plt

# 显示各年龄的人数的直方图，直方图显示的是一个数值变量的分布情况，将数据分为多个区间（也叫“桶”或“bin”），统计每个区间中的数据点数量。
# bins=100 会创建 100 个区间，因此每个区间会更窄，分布会更精细。如果减少 bins（比如 bins=20），则区间会更宽

plt.hist(df["age"], bins=100)    
plt.xlabel("age")
plt.ylabel("count")
plt.title("Histogram of ages in titanic dataset")
```

**OOP approach in matplotlib**

```python
# fig, ax = plt.subplots(2, 2)
fig, ax = plt.subplots(1)
ax.hist(df["age"], bins = 50)
ax.set(title="Histogram of ages in titanic dataset", xlabel="age", ylabel="count")
```

```python
sns.histplot(df, x = "age", hue="sex")

#  hue="sex" ： 男女分开画图
```

```python
# 直方图（Histogram）, 核密度估计图（KDE）Kernel Density Estimation）
#核密度估计是一种平滑的概率密度函数，用曲线表示变量的分布，而不是用离散的柱形图。
# 加上 kde=True 参数后，会在直方图上叠加一条平滑的分布曲线，表示数据的密度分布.
# 设置 kde=False 的效果是 只绘制直方图，不显示核密度估计 (KDE) 曲线。
# hue="sex": 按照乘客的性别（sex 列）分组，并用不同的颜色显示每个组的数据。男性和女性乘客的年龄分布将以不同颜色的直方图和 KDE 曲线呈现

sns.histplot(data = df, x = 'age', kde=True, hue="sex")
```

- 灰色部分其实是默认的柱状图叠加效果导致的，它表示 两个类别（男和女）柱子的重叠区域。这种效果是 seaborn.histplot() 的默认行为，因为它的柱状图在堆叠时并没有设置透明度或分层方式，从而在两个类别的数据重叠的地方显示为灰色。

```python
# - 解决方案：调整柱状图的堆叠模式，可以使用 multiple="dodge" 参数，让柱状图的柱子并排显示，而不是默认的堆叠形式
# multiple="dodge"：表示分组后直方图的条形在 x 轴上并排显示（而不是堆叠）。

sns.histplot(data=df, x='age', kde=True, hue="sex", multiple="dodge")
plt.show()
```
